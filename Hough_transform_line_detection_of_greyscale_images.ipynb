### Description : This python notebook contains the line detection of 3 sample invoice images using Hough transform technique.In this method, first we convert the BGR image to greyscale image and then we achieve line detection of greyscale image by converting it to binary image.

```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
from PIL import Image
```

#### 1. sample1 greyscale-binary-line detection

```python
img1=cv2.imread('inv20.jpg')
grey_img1 = cv2.cvtColor(img1, cv2.COLOR_BGR2GRAY)
ret,thresh1 = cv2.threshold(grey_img1,110,255,cv2.THRESH_BINARY)
org_img1 = Image.fromarray(thresh1)
org_img1
```

```python
edges1 = cv2.Canny(thresh1, 30, 90)
lines1 = cv2.HoughLinesP(edges1, 1, np.pi/180, 60, 50, 5)
for line in lines1:
    for x1, y1, x2, y2 in line:
        res1=cv2.line(thresh1, (x1, y1), (x2, y2), (0,0,0), 3)
result1 = Image.fromarray(res1)
result1
```
