```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
img = cv2.imread('inv10.jpg')
```

#### Note : Here, we apply Canny edge detection algorithm to the edges present in the image,25 and 300 are the thresh1 and thresh2 values respectively

```python
edges1 = cv2.Canny(img,25,300,L2gradient=True)
```

#### Output image after line detection

```python
plt.imshow(edges1,cmap='gray')
plt.show()
```
