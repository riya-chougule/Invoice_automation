### 1. Simple threshold

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

```python
img = cv2.imread('inv8.jpg',0)
ret,thresh1 = cv2.threshold(img,127,255,cv2.THRESH_BINARY)
ret,thresh2 = cv2.threshold(img,127,255,cv2.THRESH_BINARY_INV)
ret,thresh3 = cv2.threshold(img,127,255,cv2.THRESH_TRUNC)
ret,thresh4 = cv2.threshold(img,127,255,cv2.THRESH_TOZERO)
ret,thresh5 = cv2.threshold(img,127,255,cv2.THRESH_TOZERO_INV)
```

```python
titles = ['Original Image','BINARY','BINARY_INV','TRUNC','TOZERO','TOZERO_INV']
images = [img, thresh1, thresh2, thresh3, thresh4, thresh5]
```

```python
for x in range(6):
    plt.subplot(2,3,x+1),plt.imshow(images[x],'gray')
    plt.title(titles[x])
    plt.xticks([]),plt.yticks([])
```

```python
titles = ['Original Image']
images = [img]
```

```python
for x in range(1):
    plt.subplot(1,1,x+1),plt.imshow(images[x],'gray')
    plt.title(titles[x])
    plt.xticks([]),plt.yticks([])
```

```python
titles = ['BINARY']
images = [thresh1]
```

```python
for x in range(1):
    plt.subplot(1,1,x+1),plt.imshow(images[x],'gray')
    plt.title(titles[x])
    plt.xticks([]),plt.yticks([])
```

### 2. Adaptive threshold

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

```python
img = cv2.imread('inv8.jpg',0)
img = cv2.medianBlur(img,5)
```

```python
ret,th1 = cv2.threshold(img,127,255,cv2.THRESH_BINARY)
```

```python
titles = ['Original Image']
images = [img]
```

```python
for x in range(1):
    plt.subplot(1,1,x+1),plt.imshow(images[x],'gray')
    plt.title(titles[x])
    plt.xticks([]),plt.yticks([])
```

```python
titles = ['Adaptive Mean Thresholding']
images = [th1]
```

```python
for x in range(1):
    plt.subplot(1,1,x+1),plt.imshow(images[x],'gray')
    plt.title(titles[x])
    plt.xticks([]),plt.yticks([])
```

### 3. Otsu threshold

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

```python
img = cv2.imread('inv8.jpg',0)
```

```python
ret2,th2 = cv2.threshold(img,0,255,cv2.THRESH_BINARY+cv2.THRESH_OTSU)
```

```python
images = [img]
titles = ['Original Noisy Image']
```

```python
for i in range(1):
    plt.subplot(1,2,i*3+1),plt.imshow(images[i*3],'gray')
    plt.title(titles[i*3]), plt.xticks([]), plt.yticks([])
plt.show()
```

```python
images = [th2]
titles = ['Otsu Thresholding']
```

```python
for i in range(1):
    plt.subplot(1,2,i*3+1),plt.imshow(images[i*3],'gray')
    plt.title(titles[i*3]), plt.xticks([]), plt.yticks([])
plt.show()
```
