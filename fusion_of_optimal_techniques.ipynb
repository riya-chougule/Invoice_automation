### First, binarize the given image using adaptive threshold technique,with threshold value 127.

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

```python
img = cv2.imread('inv8.jpg',0)
img = cv2.medianBlur(img,5)
```

```python
ret,th1 = cv2.threshold(img,127,255,cv2.THRESH_BINARY)
th2 = cv2.adaptiveThreshold(img,255,cv2.ADAPTIVE_THRESH_MEAN_C,\
            cv2.THRESH_BINARY,11,2)
```

```python
title = ['Original Image','Adaptive Mean Thresholding']
image = [img,th2]
```

```python
for x in range(1):
    plt.subplot(1,1,x+1),plt.imshow(image[x],'gray')
    plt.title(title[x])
    plt.xticks([]),plt.yticks([])
```

```python
title2 = ['Adaptive Mean Thresholding']
image2 = [th2]
```

### Here we got a binarized image

```python
for x in range(1):
    plt.subplot(1,1,x+1),plt.imshow(image2[x],'gray')
    plt.title(title2[x])
    plt.xticks([]),plt.yticks([])
```

### Now we will blur the image using Gaussian filtering

```python
img = cv2.imread('thresh_img1.png')
```

```python
blur = cv2.GaussianBlur(img,(5,5),0)
```

```python
plt.subplot(121),plt.imshow(img),plt.title('Original')
plt.xticks([]), plt.yticks([])
plt.subplot(122),plt.imshow(blur),plt.title('Blurred')
plt.xticks([]), plt.yticks([])
plt.show()
```
