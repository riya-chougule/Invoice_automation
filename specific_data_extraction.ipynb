####                        1.Name of the vendor
####                        2.DL number of the vendor

```python
import re
import pytesseract as tess
from PIL import Image
import matplotlib.pyplot as plt
```

```python
img=Image.open("inv20.jpg")
```

```python
imgplot=plt.imshow(img)
```

```python
text=tess.image_to_string(img)
```

```python
print(text)
```

#### Note: Here we create a regular expression for obtaining the name of the vendor

```python
vendor_name_re = re.compile(r"^.*MEDICAL*")
```

```python
for line in text.split('\n'):
    if vendor_name_re.match(line):
        print(line)
```

#### Note: Here we create a regular expression for obtaining the DL No. of the vendor

```python
vendor_number_re=re.compile(r"^DL NO.*")
```

```python
for line in text.split('\n'):
    if vendor_number_re.match(line):
        print(line)
```
