```python
import cv2
import numpy as np
from skimage import io
from matplotlib import pyplot as plt
```

```python
img = cv2.cvtColor(io.imread('image.jpg'), cv2.COLOR_RGB2GRAY)
```

```python
imgplot = plt.imshow(img)
```

```python
img = cv2.threshold(img, 128, 255, cv2.THRESH_BINARY)[1]
```

```python
horizontal_size = 100
vertical_size = 100
horizontalStructure = cv2.getStructuringElement(cv2.MORPH_RECT, (horizontal_size, 1))
verticalStructure = cv2.getStructuringElement(cv2.MORPH_RECT, (1, vertical_size))
```

 ### Note : We will use two variables, mask1 and mask2 to store the horizontal and vertical lines respectively

```python
mask1 = cv2.morphologyEx(img, cv2.MORPH_OPEN, horizontalStructure)
mask2 = cv2.morphologyEx(img, cv2.MORPH_OPEN, verticalStructure)
```

### Here we will plot the results

### Extracted vertical lines from input image

```python
imgplot = plt.imshow(mask2)
```

### Extracted horizontal lines from input image

```python
imgplot = plt.imshow(mask1)
```
