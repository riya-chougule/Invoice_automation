### 1. Lines detection using Canny edge detection method

```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
img1 = cv2.imread('inv10.jpg')
edges1 = cv2.Canny(img1,25,255,L2gradient=True)
NUM_ROWS = 1
IMGs_IN_ROW = 2
f, ax = plt.subplots(NUM_ROWS, IMGs_IN_ROW, figsize=(16,6))

ax[0].imshow(img1)
ax[1].imshow(edges1)

ax[0].set_title('Original image')
ax[1].set_title('Edge Detected image')
```

### 2. Lines detection using Hough transform method

```python
img4 = cv2.imread("inv10.jpg")
grayscale4 = cv2.cvtColor(img4, cv2.COLOR_BGR2GRAY)
edges4 = cv2.Canny(grayscale4, 30, 100)
lines4 = cv2.HoughLinesP(edges4, 1, np.pi/180, 60, 50, 5)
for line in lines4:
    for x1, y1, x2, y2 in line:
        res4=cv2.line(img4, (x1, y1), (x2, y2), (20, 220, 20), 3)
NUM_ROWS = 1
IMGs_IN_ROW = 2
f4, ax4 = plt.subplots(NUM_ROWS, IMGs_IN_ROW, figsize=(16,6))
img4 = cv2.imread("inv10.jpg")
ax4[0].imshow(img4)
ax4[1].imshow(res4)

ax4[0].set_title('Original image')
ax4[1].set_title('Edge Detected image')
```
