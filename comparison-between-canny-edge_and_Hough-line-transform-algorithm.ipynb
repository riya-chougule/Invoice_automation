### 1. Lines detection using Canny edge detection method

```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
img = cv2.imread('inv10.jpg')
```

```python
edges = cv2.Canny(img,25,255,L2gradient=True)
```

```python
plt.imshow(edges,cmap='gray')
plt.show()
```

### 2. Lines detection using Hough transform method

```python
img1 = cv2.imread("inv10.jpg")
```

```python
grayscale = cv2.cvtColor(img1, cv2.COLOR_BGR2GRAY)
```

```python
edges1 = cv2.Canny(grayscale, 30, 100)
```

```python
lines = cv2.HoughLinesP(edges1, 1, np.pi/180, 60, np.array([]), 50, 5)
```

```python
for line in lines:
    for x1, y1, x2, y2 in line:
        cv2.line(img1, (x1, y1), (x2, y2), (20, 220, 20), 3)
```

```python
plt.imshow(img1)
plt.show()
```
