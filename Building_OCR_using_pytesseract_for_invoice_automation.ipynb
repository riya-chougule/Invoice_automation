### Description : In this notebook, I have built an OCR using pytesseract which firstly identifies the individual characters and then we cluster individual characters for the particular words and print them outside the boundary boxes.Here we have used binary images for building an OCR.

```python
from PIL import Image
import cv2
import numpy as np
import pytesseract
pytesseract.pytesseract.tesseract_cmd="C:\\Program Files\\Tesseract-OCR\\tesseract.exe"
```

```python
img=cv2.imread("inv10_binary.png")
```

#### Note : Here we store the shape of image in variables respectively

```python
hImg,wImg,_=img.shape
```

#### For character identification

```python
boxes=pytesseract.image_to_boxes(img)
for b in boxes.splitlines():
    b=b.split(' ')
    x,y,w,h=int (b[1]),int (b[2]),int (b[3]),int (b[4])
    cv2.rectangle(img,(x,hImg-y),(w,hImg-h),(0,0,255),2)
    cv2.putText(img,b[0],(x,hImg-y+24),cv2.FONT_HERSHEY_COMPLEX,1,(177,13,40),2)
cv2.imshow("Img",img)
cv2.waitKey(0)
cv2.destroyAllWindows()
image = Image.fromarray(img)
image
```

#### For clustering of a characters

```python
img2=cv2.imread("inv10_binary.png")
hImg1,wImg1,_=img2.shape
boxes1=pytesseract.image_to_data(img2)
for x1,b1 in enumerate(boxes1.splitlines()):
    if x1!=0:
        b1=b1.split()
        if len(b1)==12:
            x1,y1,w1,h1=int(b1[6]),int (b1[7]),int (b1[8]),int (b1[9])
            cv2.rectangle(img2,(x1,y1),(w1+x1,h1+y1),(0,0,255),2)
            cv2.putText(img2,b1[11],(x1,y1),cv2.FONT_HERSHEY_COMPLEX,1,(117,13,40),2)
cv2.imshow("Img2",img2)
cv2.waitKey(0)
cv2.destroyAllWindows()
image2 = Image.fromarray(img2)
image2
```
